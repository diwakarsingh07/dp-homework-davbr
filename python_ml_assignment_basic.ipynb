```

s = input("Enter a string: ")
vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for ch in s:
    if ch.isalpha():
        if ch in vowels:
            vowel_count = vowel_count + 1
        else:
            consonant_count = consonant_count + 1

print("Vowels =", vowel_count)
print("Consonants =", consonant_count)
```

```

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
d = int(input("Enter fourth number: "))

largest = a
if b > largest:
    largest = b
if c > largest:
    largest = c
if d > largest:
    largest = d

smallest = a
if b < smallest:
    smallest = b
if c < smallest:
    smallest = c
if d < smallest:
    smallest = d

print("Largest =", largest)
print("Smallest =", smallest)
```

```

n = int(input("Enter n: "))
a = 0
b = 1
count = 0

while count < n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    count = count + 1
```

```

pandals = {
    "Pandal 1": 1200,
    "Pandal 2": 850,
    "Pandal 3": 2300,
    "Pandal 4": 1750,
    "Pandal 5": 900
}

max_visitors = 0
max_pandal = ""

for p in pandals:
    if pandals[p] > max_visitors:
        max_visitors = pandals[p]
        max_pandal = p

print("Pandal with maximum visitors:", max_pandal, "=", max_visitors)

total = 0
for p in pandals:
    total = total + pandals[p]
average = total / len(pandals)
print("Average visitors =", average)
```

```

def festive_discount(price, percent):
    discount = (price * percent) / 100
    return price - discount

print(festive_discount(1000, 10))
print(festive_discount(500, 20))
print(festive_discount(250, 5))
```

```

cm = [[45, 5],
      [7, 43]]

TP = cm[0][0]
FN = cm[0][1]
FP = cm[1][0]
TN = cm[1][1]

total = TP + TN + FP + FN

accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)

print("Accuracy =", accuracy)
print("Precision =", precision)
print("Recall =", recall)
print("F1 Score =", f1)
```

```

def metrics(cm):
    TP = cm[0][0]
    FN = cm[0][1]
    FP = cm[1][0]
    TN = cm[1][1]

    total = TP + TN + FP + FN
    accuracy = (TP + TN) / total
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1 = 2 * precision * recall / (precision + recall)

    return accuracy, precision, recall, f1

print(metrics([[45, 5], [7, 43]]))
```
